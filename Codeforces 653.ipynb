```python
#A
t = int(input())

for _ in range(t):
    x, y, n = map(int,input().split())
    print(x*((n-y)//x)+y)
```

```python
#B
def solve(p):
    two = 0
    three = 0
    while p > 1:
        if p/2 != p//2:
            break
        p //= 2
        two += 1
    while p > 1:
        if p/3 != p//3:
            break
        p //= 3
        three += 1
    if p != 1:
        return -1
    if two > three:
        return -1
    return (three-two)+three

t = int(input())
for i in range(t):
    n = int(input())
    ans = solve(n)
    print(ans)
```

```python
#C
from collections import deque

t = int(input())
for _ in range(t):
    n = int(input())
    S = input()
    q = deque([])
    ans = 0
    for s in S:
        if s == "(":
            q.append(s)
        else:
            if not q:
                ans += 1
            else:
                q.popleft()
    print(ans)
```

```python
#D
t = int(input())

for _ in range(t):
    n, k = map(int,input().split())
    A = list(map(lambda x:k-int(x)%k if int(x)%k else 0, input().split()))
    
    cnt = 0
    d = {}
    for a in A:
        if a == 0:
            cnt += 1
            continue
        if a not in d:
            d[a] = 1
        else:
            d[a] += 1
    
    if not d:
        print(0)
        continue
    
    maxim = max(d.values())
    keys = [i for i in d.keys() if d[i] == maxim]
    key = max(keys)
    
    ans = k*(maxim-1)+key+1
    print(ans)
```

```python
#E
import itertools

n, k = map(int,input().split())

both = [0]
alice = [0]
bob = [0]
for _ in range(n):
    t, a, b = map(int,input().split())
    if a and b:
        both.append(t)
    elif a:
        alice.append(t)
    elif b:
        bob.append(t)
        
if len(alice)+len(both) < k or len(bob)+len(both) < k:
    print(-1)
    exit()
    
both.sort()
alice.sort()
bob.sort()

cs_both = list(itertools.accumulate(both))
cs_alice = list(itertools.accumulate(alice))
cs_bob = list(itertools.accumulate(bob))

"""
print(cs_both)
print(cs_alice)
print(cs_bob)
"""

ans = float('inf')
for i in range(k+1):
    try:
        tmp = cs_both[i]
    except IndexError:
        break
    if i == k:
        ans = min(ans, tmp)
    try:
        tmp += cs_alice[k-i]
    except IndexError:
        continue
    try:
        tmp += cs_bob[k-i]
    except IndexError:
        continue
    ans = min(ans, tmp)

if ans == float('inf'):
    print(-1)
else:
    print(ans)
```
