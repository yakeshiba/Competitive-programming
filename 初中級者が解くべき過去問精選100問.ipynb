```python
#ITP1_7_B - How Many Ways?
ns = []
xs = []
while True:
    n,x = map(int,input().split())
    if n == 0 and x == 0:
        break
    ns.append(n)
    xs.append(x)
ans = []
for i in range(len(ns)):
    tmp = 0
    for a in range(1,ns[i]+1):
        for b in range(a,ns[i]+1):
            for c in range(b,ns[i]+1):
                if a+b+c == xs[i]:
                    if a!=b and b!=c and c!= a:
                        tmp += 1
    ans.append(tmp)
print(*ans,sep="\n")
```

```python
#AtCoder Beginner Contest 106 B - 105
N = int(input())

def make_divisors(n):
    from collections import deque
    divisors = deque([])
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            divisors.append(i)
            if i != n // i:
                divisors.append(n//i)
    lst_divisors = list(divisors)
    lst_divisors.sort()
    return lst_divisors

ans = 0
for i in range(1,-(-N//2)+1):
    now = 2*i-1
    if len(make_divisors(now)) == 8:
        ans += 1

print(ans)
```

```python
#AtCoder Beginner Contest 128 C - Switches
N,M = map(int,input().split())

related_switch = [[] for _ in range(M)]
for i in range(M):
    s = list(map(lambda x: int(x)-1,input().split()))
    related_switch[i] = s[1:]

p = list(map(int,input().split()))

ans = 0
for status in range(1<<N):#スイッチのon/offの状態
    on = []#onになってるスイッチの番号
    for mask in range(N):
        if (1<<mask)&status:
            on.append(mask)
    for i,s in enumerate(related_switch):
        if len(set(s)&set(on))%2 != p[i]:#積集合
            break
    else:
        ans += 1

print(ans)
```

```python
#AtCoder Beginner Contest 002 D - 派閥
import itertools
N,M = map(int,input().split())

edge = [[0]*N for _ in range(N)]
for i in range(M):
    x,y = map(int,input().split())
    edge[x-1][y-1] = 1
    edge[y-1][x-1] = 1

ans = 1
for status in range(1<<N):#誰が所属しているかを表す状態
    group = []
    for mask in range(N):
        if (1<<mask)&status:
            group.append(mask)
    for p1, p2 in itertools.combinations(group, 2):
        if edge[p1][p2] == 0:
            break
    else:
        ans = max(ans,len(group))
print(ans)
```

```python
#JOI 2008 予選 5 - おせんべい
#numpy sum max axis
import sys
import numpy as np
import copy

input=sys.stdin.readline

def main(args):
    gyou, retsu = map(int,input().split())
    
    senbei = np.array([list(input().split()) for _ in range(gyou)]) == "1"
    
    ans = 0
    for status in range(1<<gyou):
        senbei_tmp = copy.deepcopy(senbei)
        for mask in range(gyou):
            if (1<<mask)&status:
                senbei_tmp[mask] = np.logical_not(senbei_tmp[mask])
        
        white = np.sum(senbei_tmp, axis=0)
        tmp = np.sum(np.max([white,gyou-white], axis=0))
        
        ans = max(ans,tmp)
    
    print(ans)

if __name__ == '__main__':
    main(sys.argv[1:])
```

# 全探索：順列全探索 [2/3]

```python
# AtCoder Beginner Contest 145 C - Average Length
import itertools
N = int(input())
position = []
for _ in range(N):
    position.append(list(map(int,input().split())))

ans = 0
for p in itertools.permutations([i for i in range(N)]):
    for i,v in enumerate(p):
        x,y = position[v]
        if i == 0:
            nowx = x
            nowy = y
        else:
            ans += ((x-nowx)**2+(y-nowy)**2)**0.5
            nowx = x
            nowy = y

pattern = 1
for i in range(2,N+1):
    pattern *= i

print(ans/pattern)
```

# 二分探索 [2/6]

```python
#ALDS_4_B - 二分探索
def binary_s(lst, value):
    left = 0
    right = len(lst)-1
    while left <= right:
        mid = (left + right)//2
        if lst[mid] == value:
            return mid
        elif lst[mid] < value:
            left = mid + 1
        else:
            right = mid -1
    return -1

n = int(input())
S = list(map(int,input().split()))
q = int(input())
T = list(map(int,input().split()))

ans = 0
for t in T:
    if binary_s(S, t) >= 0:
        ans += 1

print(ans)
```

```python
#AtCoder Beginner Contest 023 D - 射撃王
def is_ok(middle):
    timelimit = [0]*N
    for i in range(N):
        if middle < H[i]:
            return False
        timelimit[i] = (middle-H[i])/S[i]
    
    timelimit.sort()
    
    for i in range(N):
        if timelimit[i] < i:
            return False
    return True

def meguru_bisect(ng, ok):
    while (abs(ok - ng) > 1):
        mid = (ok + ng) // 2
        if is_ok(mid):
            ok = mid
        else:
            ng = mid
    return ok

N = int(input())
H = [0]*N
S = [0]*N
for i in range(N):
    h, s = map(int,input().split())
    H[i] = h
    S[i] = s

ans = meguru_bisect(-1, 10**18+1)

print(ans)
```

# 幅優先探索 [2/6]

```python
#ALDS_11_C - 幅優先探索
from collections import deque

N = int(input())

edge = [[] for _ in range(N)]
for _ in range(N):
    t = list(map(int,input().split()))
    if t[1] != 0:
        edge[t[0]-1] = list(map(lambda x:int(x)-1, t[2:]))

dist = [-1]*N
dist[0] = 0
        
q = deque([0])
while q:
    now = q.popleft()
    for nxt in edge[now]:
        if dist[nxt] == -1:
            dist[nxt] = dist[now] + 1
            q.append(nxt)
        else:
            dist[nxt] = min(dist[nxt], dist[now] + 1)

for i,d in enumerate(dist):
    print(i+1, d)
```

```python
#AtCoder Beginner Contest 007 C - 幅優先探索
from collections import deque

R,C = map(int,input().split())
sy, sx = map(lambda x:int(x)-1, input().split())
gy, gx = map(lambda x:int(x)-1, input().split())

maze = [0]*R
for i in range(R):
    maze[i] = list(input())
    
distance = [[float('inf')]*C for _ in range(R)]
distance[sy][sx] = 0

dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]


q = deque([(sy, sx)])
while q:
    y, x = q.popleft()
    maze[y][x] = "#"
    for dxi,dyi in zip(dx,dy):
        if not (0<=y+dyi<R and 0<=x+dxi<C) or maze[y+dyi][x+dxi] == "#":
            pass
        else:
            distance[y+dyi][x+dxi] = min(distance[y][x]+1, distance[y+dyi][x+dxi])
            maze[y+dyi][x+dxi] = "#"
            q.append((y+dyi, x+dxi))
print(distance[gy][gx])
```

```python
#DPL_1_B - 0,1ナップザック問題
#参考URL:https://mathwords.net/napusaku
n, w = map(int,input().split())

V = []
W = []
for i in range(n):
    v, ww = map(int,input().split())
    V.append(v)
    W.append(ww)
    
dp = [[0]*(w+1) for _ in range(n)]

for i in range(n):# i番目の商品までで考えたときの最大値
    for wgt in range(1,w+1):
        if wgt-W[i] < 0:
            dp[i][wgt] = dp[i-1][wgt]
            continue
        dp[i][wgt] = max(dp[i-1][wgt], dp[i-1][wgt-W[i]]+V[i])

print(max(dp[n-1]))
```

```python
#DPL_1_C - ナップザック問題
N,W = map(int,input().split())

product = [(0,0)]
for _ in range(N):
    v,w = map(int,input().split())
    product.append((w, v))

#縦が商品i，横が仮想カバンの耐久値
#i番目までの商品だけを使って，耐久wのカバンの価値を最大化していく
dp = [[-float('inf')]*(W+1) for _ in range(N+1)]

for i in range(N+1):
    for w in range(W+1):
        if i == 0:
            dp[i][w] = 0
            continue
        if w == 0:
            dp[i][w] = 0
            continue
        if product[i][0] > w:
            dp[i][w] = dp[i-1][w]
        else:
            not_choice = dp[i-1][w]
            choice = dp[i-1][w-product[i][0]] + product[i][1]
            same_choice = dp[i][w-product[i][0]] + product[i][1]#重複を許して選ぶことに相当
            dp[i][w] = max(not_choice, choice, same_choice)
print(max(dp[N]))
```

# ダイクストラ法 [1/4]
### N頂点 M辺のグラフにおける頂点 rから各頂点への最短経路長を、O(MlogN)で計算するアルゴリズム
### いつもの距離求解にheapqを使っているだけ

```python
#GRL_1_A - 単一始点最短経路
from heapq import heappush, heappop

v,e,r = map(int,input().split())

edge = [[] for _ in range(v)]

for _ in range(e):
    s,t,d = map(int,input().split())
    edge[s].append((t, d))

cost = [float('inf')]*v
cost[r] = 0

q = [(r, 0)]
while q:
    now, distance = heappop(q)
    #print(q,now,distance)
    
    for nxt, newcost in edge[now]:
        if distance + newcost < cost[nxt] :
            cost[nxt] = distance + newcost
            heappush(q, (nxt, distance + newcost))

for i in range(v):
    if cost[i] == float('inf'):
        print('INF')
    else:
        print(cost[i])
```

# ワ―シャルフロイド法 [3/4]
### N頂点 M辺のグラフにおける全頂点対間の最短経路長を、O(N^3)で計算するアルゴリズム

```python
#GRL_1_C - 全点対間最短経路
#有向グラフのとき
def warshallFloyd(V, dp):
    for k in range(V):
        for i in range(V):
            for j in range(V):
                dp[i][j] = min(dp[i][j], dp[i][k] + dp[k][j])

V, E = map(int,input().split())
dp = [[float('inf')]*V for _ in range(V)]
for i in range(V):
    dp[i][i] = 0
for _ in range(E):
    f, t, d = map(int,input().split())
    dp[f][t] = d
warshallFloyd(V, dp)

ncircle = False
for i in range(V):
    if dp[i][i] < 0:#負経路＝ある点から出て自分に帰ってくる経路＝点iから点i
        ncircle = True
if ncircle:
    print('NEGATIVE CYCLE')
else:
    for d in dp:
        print(' '.join(map(str, d)).replace('inf', 'INF'))
```

# 逆元を使う問題 [1/4]

```python
#AtCoder Beginner Contest 034 C - 経路
mod = 10**9+7

def modInverse(a,p):
    # Fermat's little theorem, a**(p-1) = 1 mod p
    return pow(a,p-2,p)

def nCr(n,r,p=mod):
    num = 1
    den = 1
    for i in range(r):
        num = (num*(n-i)) % p
        den = (den*(i+1)) % p
    return num*modInverse(den,p)% p
    
W,H = map(int,input().split())
print(nCr(W+H-2,H-1))
```

```python
#DSL_1_A - 互いに素な集合
class UnionFind:
    def __init__(self, n):
        self.n = n
        self.parents = [-1] * n
 
    def find(self, x):#要素xが属するグループの根を返す
        if self.parents[x] < 0:
            return x
        else:
            self.parents[x] = self.find(self.parents[x])
            return self.parents[x]
 
    def union(self, x, y):#要素xが属するグループと要素yが属するグループとを併合する
        x = self.find(x)
        y = self.find(y)
 
        if x == y:
            return
 
        if self.parents[x] > self.parents[y]:
            x, y = y, x
 
        self.parents[x] += self.parents[y]
        self.parents[y] = x
 
    def size(self, x):#要素xが属するグループのサイズ（要素数）を返す
        return -self.parents[self.find(x)]
 
    def same(self, x, y):#要素x, yが同じグループに属するかどうかを返す
        return self.find(x) == self.find(y)
 
    def members(self, x):#要素xが属するグループに属する要素をリストで返す
        root = self.find(x)
        return [i for i in range(self.n) if self.find(i) == root]
 
    def roots(self):#すべての根の要素をリストで返す
        return [i for i, x in enumerate(self.parents) if x < 0]
 
    def group_count(self):#グループの数を返す
        return len(self.roots())
 
    def all_group_members(self):#{ルート要素: [そのグループに含まれる要素のリスト], ...}の辞書を返す
        return {r: self.members(r) for r in self.roots()}
 
    def __str__(self):#print()での表示用 ルート要素: [そのグループに含まれる要素のリスト]を文字列で返す
        return '\n'.join('{}: {}'.format(r, self.members(r)) for r in self.roots())

n,q = map(int,input().split())

uf = UnionFind(n)

for i in range(q):
    query = list(map(int,input().split()))
    if query[0] == 0:
        uf.union(query[1], query[2])
    else:
        print(int(uf.same(query[1], query[2])))
```
