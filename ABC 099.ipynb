```python
#C
N = int(input())
nine = []
six = []
ans = 0
i = 1
while True:
    if 9**i > N:
        break
    else:
        nine.append(9**i)
        i += 1
i = 1
while True:
    if 6**i > N:
        break
    else:
        six.append(6**i)
        i += 1

print(nine)
print(six)
print("##########")

if len(six) == 0 and len(nine) == 0:
    ans = N
else:
    while True:
        if len(six) != 0 and len(nine) != 0:
            while nine[-1] > N and len(nine) > 0:
                nine.pop(-1)
            while six[-1] > N and len(six) > 0:
                six.pop(-1)
            withdraw = max(six[-1],nine[-1])
        if len(nine) == 0:
            while six[-1] > N and len(six) > 0:
                six.pop(-1)
            withdraw = six[-1]
        if len(six) == 0:
            while nine[-1] > N and len(nine) > 0:
                nine.pop(-1)
            withdraw = nine[-1]
        N -= withdraw
        if N < 0:
            #残りは１円で
            ans += (N+withdraw)
            break
        else:
            ans += 1
            if N - withdraw < 0:#次も同額引き出せないとき
                if len(nine) != 0 and withdraw == nine[-1]:#9のべき乗額下ろしてたら
                    nine.pop(-1)
                elif len(six) != 0 and withdraw == six[-1]:
                    six.pop(-1)
        print("withdraw",withdraw)
        print(nine)
        print(six)
        print("nokori",N,"YEN")
        print("count",ans)
        print("####")
        if N == 0:
            break
print(ans)
```

```python
N = int(input())
nine = []
six = []
ans = 0
i = 1
while True:
    if 9**i > N:
        break
    else:
        nine.append(9**i)
        i += 1
i = 1
while True:
    if 6**i > N:
        break
    else:
        six.append(6**i)
        i += 1

if len(six) == 0 and len(nine) == 0:
    ans = N
else:
    while True:
        if len(nine) != 0:
            while nine[-1] > N:
                nine.pop(-1)
                if len(nine) == 0:
                    break
        if len(six) != 0:
            while six[-1] > N:
                six.pop(-1)
                if len(six) == 0:
                    break
        if len(nine) != 0 and len(six) != 0:
            withdraw = max(nine[-1],six[-1])
        elif len(nine) == 0 and len(six) != 0:
            withdraw = six[-1]
        elif len(six) == 0 and len(nine) != 0:
            withdraw = nine[-1]
        else:
            ans += N
            break
        print(nine)
        print(six)
        print("N",N)
        count = N//withdraw
        N -= withdraw*count
        print("withdraw",withdraw)
        print("withdraw*count",withdraw*count)
        print("nokori",N)
        ans += count
        print("ans",ans)
        print("###########")
print(ans)
```
