```python
#D
from collections import defaultdict

S = input()
d = defaultdict(int)
for s in S:
    d[s] += 1

if len(S) == 1:
    if S[0] == "8":
        print('Yes')
    else:
        print('No')
    exit()
elif len(S) == 2:
    for i in range(10,99):
        if i%8 != 0:
            continue
        s1 = str(i)[0]
        s2 = str(i)[1]
        
        num = [0]*10
        for j in range(1,10):
            num[j] = d[str(j)]
        num[int(s1)] -= 1
        num[int(s2)] -= 1
        
        f = False
        for j in range(0,10):
            if num[j] < 0:
                f = True
                break
        if f:
            continue
        print('Yes')
        exit()
        
    print('No')
    
else:
    for i in range(100,1000):
        if i%8 != 0:
            continue
        s1 = str(i)[0]
        s2 = str(i)[1]
        s3 = str(i)[2]

        num = [0]*10
        for j in range(1,10):
            num[j] = d[str(j)]
        num[int(s1)] -= 1
        num[int(s2)] -= 1
        num[int(s3)] -= 1
        
        f = False
        for j in range(0,10):
            if num[j] < 0:
                f = True
                break
        if f:
            continue
        print('Yes')
        exit()
        
        
    print('No')
```

```python
#E
import bisect

class SegmentTree:
    # 初期化処理
    # f : SegmentTreeにのせるモノイド
    # default : fに対する単位元
    def __init__(self, size, f=lambda x,y : x+y, default=0):
        self.size = 2**(size-1).bit_length() # 簡単のため要素数Nを2冪にする
        self.default = default
        self.dat = [default]*(self.size*2) # 要素を単位元で初期化
        self.f = f

    def update(self, i, x):
        i += self.size
        self.dat[i] = x
        while i > 0:
            i >>= 1
            self.dat[i] = self.f(self.dat[i*2], self.dat[i*2+1])

    def query(self, l, r):
        l += self.size
        r += self.size
        lres, rres = self.default, self.default
        while l < r:
            if l & 1:
                lres = self.f(lres, self.dat[l])
                l += 1

            if r & 1:
                r -= 1
                rres = self.f(self.dat[r], rres) # モノイドでは可換律は保証されていないので演算の方向に注意
            l >>= 1
            r >>= 1
        res = self.f(lres, rres)
        return res

n, m = map(int,input().split())
H = list(map(int,input().split()))
W = list(map(int,input().split()))
H.sort()

pair = 0
sum_0 = []
sum_1 = []
for i in range(0,n,2):
    try:
        sum_0.append(abs(H[i+1]-H[i]))
    except IndexError:
        break
for i in range(1,n,2):
    try:
        sum_1.append(abs(H[i+1]-H[i]))
    except IndexError:
        break

# min,math.gcd など，関数を入力することでqueryで行う演算を指定
# minのときは初期化が必要（他にも必要な関数があるかも）s = SegmentTree(n, min, 2**31-1)
# 区間和，区間積は lambda a,b:a+b や lambda a,b:a*b で指定
seg_sum0 = SegmentTree(n, lambda a,b:a+b)
seg_sum1 = SegmentTree(n, lambda a,b:a+b)

#s[i]をxに置換
for i,x in enumerate(sum_0):
    seg_sum0.update(i, x)
for i,x in enumerate(sum_1):
    seg_sum1.update(i, x)
    
ans = float('inf')
for w in W:
    ind = bisect.bisect_left(H, w)
    if ind%2 == 1:
        tmp = abs(w-H[ind-1])
    else:
        tmp = abs(w-H[ind])
    tmp += seg_sum0.query(0, ind//2)
    tmp += seg_sum1.query(ind//2, len(sum_1))
    ans = min(tmp, ans)
    
print(ans)
```
