```python
#C
from collections import deque

def check753(num):
    three = False
    five = False
    seven = False
    for s in num:
        if s == "3":
            three = True
        elif s == "5":
            five = True
        elif s == "7":
            seven = True
        if three and five and seven:
            return True
    return False

n = int(input())

q = deque([3,5,7])

ans = 0
while q:
    now = q.popleft()
    
    t = "".join([str(now), "3"])
    if int(t) <= n:
        q.append(int(t))
        if check753(t):
            ans += 1
    f = "".join([str(now), "5"])
    if int(f) <= n:
        q.append(int(f))
        if check753(f):
            ans += 1
    s = "".join([str(now), "7"])
    if int(s) <= n:
        q.append(s)
        if check753(s):
            ans += 1
print(ans)
```

```python
N = int(input())

def factorization(n):
    arr = []
    temp = n
    for i in range(2, int(-(-n**0.5//1))+1):
        if temp%i==0:
            cnt=0
            while temp%i==0:
                cnt+=1
                temp //= i
            arr.append([i, cnt])

    if temp!=1:
        arr.append([temp, 1])

    if arr==[]:
        arr.append([n, 1])

    return arr

factor = {}
for i in range(2,N+1):
    tmp = factorization(i)
    #print(tmp)
    for s in tmp:
        if s[0] not in factor:
            factor[s[0]] = s[1]
        else:
            factor[s[0]] += s[1]
#print(factor)
ans = 0
two_three = 0
geq4 = 0
geq14 = 0
geq24 = 0
geq74 = 0
for v in factor.values():
    if v == 2 or v == 3:
        two_three += 1
    if v >= 4:
        geq4 += 1
    if v >= 14:
        geq14 += 1
    if v >= 24:
        geq24 += 1
    if v >= 74:
        geq74 += 1
#2,4,4
p1 = two_three*(geq4)*(geq4-1)//2
p2 = geq4*(geq4-1)*(geq4-2)//6
#2,24
q1 = (two_three+geq4-geq24)*(geq24)
q2 = geq24*(geq24-1)//2
#4,14
r1 = (geq4-geq14)*(geq14)
r2 = geq14*(geq14-1)//2
print(p1+p2+q1+q2+r1+r2+geq74)
```
