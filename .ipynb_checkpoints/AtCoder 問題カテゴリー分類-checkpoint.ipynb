```python
#ABC087C - Candies
import itertools

N = int(input())

A1 = list(map(int,input().split()))
A2 = list(map(int,input().split()))

#どこで下の行に移るか
csA1 = list(itertools.accumulate(A1))
csA2 = list(itertools.accumulate(A2))

#print(csA1)
#print(csA2)

ans = A1[0] + csA2[-1]
for i in range(N-1):
    tmp = csA1[i] + (csA2[-1] - csA2[i]) + A2[i]
    ans = max(ans,tmp)
print(ans)
```

```python
#ABC098C - Attention
#ある一点を基準に，その左右の最小値を求める
import numpy as np
N = int(input())
S = "d"+input()+"d"

#index=0,N+1は初期化に必要
leftwest = np.array([0]*(N+2))
righteast = np.array([0]*(N+2))

for i in range(1,N+1):
    if S[i] == "W":
        leftwest[i+1] = leftwest[i] + 1
    else:
        leftwest[i+1] = leftwest[i]
        
    if S[-(i+1)] == "E":
        righteast[-(i+2)] = righteast[-(i+1)] + 1
    else:
        righteast[-(i+2)] = righteast[-(i+1)]

#print("left",leftwest)
#print('right',righteast)


print(min(leftwest[1:N+1]+righteast[1:N+1]))
```

```python
#ABC122C - GeT AC
N, Q = map(int,input().split())
S = input()+"d"
#'AC'が見つかったらCのほうに+1
count = [0]*(N+1)
for i in range(N):
    if S[i] == "A":
        if S[i+1] == "C":
            count[i+1] += count[i] + 1
        else:
            count[i+1] = count[i]
    else:
        count[i+1] = count[i]
    
#print(count)

for _ in range(Q):
    l, r = map(int,input().split())
    print(count[r-1]-count[l-1])
```

```python
#ABC125C - GCD on Blackboard
#「ある一点を基準に，その点以外の最大公約数を求める」×N点分
from math import gcd as gcd
import numpy as np
N = int(input())
A = list(map(int,input().split()))

A = [0] + A + [0]

fromleft = np.array([0]*N)
fromright = np.array([0]*N)

for i in range(1,N):
    fromleft[i] = gcd(A[i],fromleft[i-1])
    fromright[-(i+1)] = gcd(A[-(i+1)],fromright[-i])

#print(fromleft)
#print(fromright)

ansgcd = [0]*N

for i in range(N):
    ansgcd[i] = gcd(fromright[i], fromleft[i])

print(max(ansgcd))
```

```python
#ABC084D - 2017-like Number
def create_prime_list(limit):
    x = limit**0.5
    primes = []
    #print('x={0}'.format(x))
    nums = [x for x in range(2, limit+1)]
    while nums[0]<=x:
        primes.append(nums[0])
        current_prime = nums[0]
        nums = [x for x in nums if x%current_prime != 0]
    primes.extend(nums)
    return primes
    
Q = int(input())

p = set(create_prime_list(10**5))

#今注目してる数より左に出てきた「2017に似た数」
n_of_p = [0]*(10**5+2)

for i in range(1,10**5+1):
    if i%2 == 1:
        if i in p and (i+1)//2 in p:
            n_of_p[i+1] = n_of_p[i] + 1
        else:
            n_of_p[i+1] = n_of_p[i]
    else:
        n_of_p[i+1] = n_of_p[i]
        
#print(n_of_p[0:20])

for _ in range(Q):
    l,r = map(int,input().split())
    print(n_of_p[r+1] - n_of_p[l])
```

```python
#ABC089D - Practical Skill Test
H,W,D = map(int,input().split())

A = [[] for _ in range(H)]
Aset = [[] for _ in range(H)]
for i in range(H):
    A[i] = list(map(int,input().split()))
    Aset[i] = set(A[i])

#1,1+D,1+2D,...
#2,2+D,2+2D,...
#...
#D,2D,3D,...
cost = [0]*(H*W+1)
for i in range(1,D+1):
    num = i
    now = None#座標
    while True:
        if num > H*W:
            break
        for row in range(H):
            if num in Aset[row]:
                #print(row)
                ind = A[row].index(num)
                if now == None:
                    now = (ind, row)#x,y
                    cost[num] = 0
                else:
                    cost[num] = cost[num-D] + abs(ind-now[0]) + abs(row-now[1])
                    now = (ind, row)
        num += D

Q = int(input())
for _ in range(Q):
    l, r = map(int,input().split())
    print(cost[r]-cost[l])
```

```python
#ABC129D - Lamp
import numpy as np
H ,W = map(int,input().split())

maze = [[] for _ in range(H)]
for i in range(H):
    maze[i] = np.array(list(input())) == "."
maze = np.array(maze)

up = np.zeros((H, W), dtype=int)
down = np.zeros((H, W), dtype=int)
right = np.zeros((H, W), dtype=int)
left = np.zeros((H, W), dtype=int)

for i in range(H):
    up[i] = (up[i-1] + 1)*maze[i]
    down[-(i+1)] = (down[-i] + 1)*maze[-(i+1)] 
    
#print(up)
#print(down)

for i in range(W):
    right[:, i] = (right[:, i-1] + 1)*maze[:, i]
    left[:, -(i+1)] = (left[:, -i] + 1)*maze[:, -(i+1)]

print(np.max(up+down+right+left)-3)
```

```python
#ABC038C - 単調増加
N = int(input())
A = list(map(int,input().split()))

l = [0]*(N+1)

prev = 0
for i,a in enumerate(A):
    if prev < a:
        l[i+1] = l[i] + 1
    else:
        l[i+1] = 1
    prev = a
    
print(sum(l))
```

```python
#ABC014C - AtColor
import itertools
import sys
input=sys.stdin.readline

n = int(input())

color = [0]*(10**6+1)

for _ in range(n):
    a,b = map(int,input().split())
    color[a] += 1
    if b+1 < 10**6+1:
        color[b+1] -= 1

print(max(list(itertools.accumulate(color))))
```

```python
#ABC017C - ハイスコア
import itertools
import numpy as np

N,M = map(int,input().split())

jwl = np.array([0]*M)
sum_of_val = 0 
for _ in range(N):#sがその遺跡で得られる点
    l, r, s = map(int,input().split())
    sum_of_val += s
    jwl[l-1] -= s
    if r < M:
        jwl[r] += s

cumsum = np.cumsum(jwl)

ans = 0
for i in range(M):
    ans = max(ans,sum_of_val+cumsum[i])

print(ans)
```

```python
#ABC080D - Recording
import numpy as np
import sys
input = sys.stdin.readline

def main(args):
    N,C = map(int,input().split())
    
    time = [[0]*(10**5+1) for _ in range(C)]
    
    for _ in range(N):
        s, t, c = map(lambda x:int(x)-1,input().split())
        time[c-1][s] += 1
        time[c-1][t+1] -= 1
     
    #正の値を1,0を0として配列を上書き
    recorder = np.where(np.cumsum(time[0]) > 0, 1, 0)
    
    for i in range(1,C):
        recorder += np.where(np.cumsum(time[i]) > 0, 1, 0)
    
    print(max(recorder))
    
if __name__ == '__main__':
    main(sys.argv[1:])
```

# SlidingWindow

```python
#ABC124D - Handstand
N, K = map(int,input().split())
S = input()

#圧縮（正の数が直立，負の数が逆立ち）
people = []
tmp = 0
for s in S:
    if s == "0":#直立
        if tmp >= 0:
            tmp += 1
        else:
            people.append(tmp)
            tmp = 1
    else:#逆立ち
        if tmp <= 0:
            tmp -= 1
        else:
            people.append(tmp)
            tmp = -1
people.append(tmp)

#print(people)

rev = 0#反転回数
ans = 0
#初めの範囲で
for i,num in enumerate(people):
    if num < 0:#元から逆立ちの時
        ans += -num
    else:
        rev += 1
        ans += num
    if rev == K:#それ以上反転できないので，右隣の逆立ち(居れば)の分だけ足して終わる
        try:
            ans += -people[i+1]
            break
        except IndexError:#最後に足した逆立ち群がそもそも右端だったら
            break

#print("initial ans",ans)

#SlidingWindow 右端と左端
right = (i+1)+1
left = 0

tmp = ans
#SlidingWindow
while True:
    #左端の更新
    if people[left] < 0:
        tmp -= abs(people[left])
        try:
            tmp -= abs(people[left+1])
            left += 2
        except IndexError:#len(people)==1のときだけ
            break
    else:
        tmp -= abs(people[left])
        left += 1
    
    #右端の更新
    try:
        tmp += abs(people[right])
        right += 1
    except IndexError:
        break
    
    try:
        tmp += abs(people[right])
        right += 1
    except IndexError:
        break
    ans = max(tmp, ans)

ans = max(tmp, ans)
print(ans)
```

# DFS
### 再帰関数を使うときはPythonを使う（PyPyは激遅）

```python
#ATC001A - 深さ優先探索
import sys
sys.setrecursionlimit(10**9)
input = sys.stdin.readline

def dfs(x, y, W, H, maze, visited):
    if not (0<=x<W and 0<=y<H) or maze[y][x] == "#":
        return 0
    if visited[y][x]:
        return 0
    
    visited[y][x] = 1
    
    dfs(x+1, y, W, H, maze, visited)
    dfs(x-1, y, W, H, maze, visited)
    dfs(x, y+1, W, H, maze, visited)
    dfs(x, y-1, W, H, maze, visited)
    
    return

def main(args):
    H,W = map(int,input().split())
    
    
    maze = [[] for _ in range(H)]
    for i in range(H):
        maze[i] = list(input())
        for j in range(W):
            if maze[i][j] == "s":
                sx,sy = j, i#x,y
            if maze[i][j] == "g":
                gx, gy = j, i
    
    visited = [[0]*W for _ in range(H)]
    
    dfs(sx, sy, W, H, maze, visited)
    if visited[gy][gx]:
        print('Yes')
    else:
        print('No')

if __name__ == '__main__':
    main(sys.argv[1:])
```

```python
#ABC070D - Transit Tree Path
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10**9)

def dfs(now, vis, dis):
    for to,c in edge[now]:
        if not vis[to]:
            if dis[to] == 0:
                dis[to] = dis[now] + c
            else:
                dis[to] = min(dis[now] + c, dis[to])
            vis[now] = 1
            dfs(to, vis, dis)

N = int(input())

edge = [[] for _ in range(N)]
for _ in range(N-1):
    a, b, c = map(int, input().split())
    edge[a-1].append([b-1, c])
    edge[b-1].append([a-1, c])

visited = [0]*N
distance = [0]*N

Q, K = map(int, input().split())
dfs(K-1, visited, distance)

#print(distance)

for _ in range(Q):
    x, y = map(lambda x:int(x)-1, input().split())
    print(distance[x]+distance[y])
```

# 二分探索

```python
#ABC077C - Snuke Festival
import bisect

N = int(input())
A = list(map(int,input().split()))
B = list(map(int,input().split()))
C = list(map(int,input().split()))

A.sort()
B.sort()
C.sort()

#print(A)
#print(B)
#print(C)

ans = 0
for b in B:
    a = bisect.bisect_right(A, b-1)
    c = N - bisect.bisect_right(C, b)
    #print("smaller",a,"bigger",c)
    ans += a*c

print(ans)
```
