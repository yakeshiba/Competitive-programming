```python
#A
from collections import Counter

n = int(input())
A = list(map(int,input().split()))

cnt = Counter(A)

app = [0]*(max(A)+1)
ans = []
for a in A:
    app[a] += 1
    if app[a] == cnt[a]:
        ans.append(a)

print(len(ans))
print(*ans)
```

```python
#C
import itertools
import bisect

n, m = map(int,input().split())
A = list(map(int,input().split()))
B = list(map(int,input().split()))

csA = [0]+list(itertools.accumulate(A))
#print(csA)

ans = []
for b in B:
    ind = bisect.bisect_left(csA, b)
    ans.append((ind, b-csA[ind-1]))
    
for a, b in ans:
    print(a, b)
```

```python
#D
import copy

n = int(input())
Bori = list(map(int,input().split()))

if n <= 2:
    print(0)
    exit()

ans = n+1
op = [-1, 0, 1]
for i in op:
    for j in op:
        B = copy.deepcopy(Bori)
        tmp = abs(i)+abs(j)
        
        b0 = B[0] + i
        b1 = B[1] + j
        
        diff = b1 - b0
        B[1] += j
        for b in range(2,n):
            if B[b] - B[b-1] == diff:
                continue
            elif B[b] - B[b-1] == diff+1:
                tmp += 1
                B[b] -= 1
            elif B[b] - B[b-1] == diff-1:
                tmp += 1
                B[b] += 1
            else:
                break
        else:
            ans = min(ans, tmp)
            
if ans == n+1:
    print(-1)
else:
    print(ans)
```

```python
#E
import itertools

n, w = map(int,input().split())
A = list(map(int,input().split()))

csA = list(itertools.accumulate(A))
#print(csA)

if max(csA) > w:
    print(0)
    exit()
    
low = -min(0,min(csA))
high = min(w, w-max(csA))

print(high-low+1 if low <= high else 0)
```

```python
#F
from collections import Counter

n, k = map(int,input().split())
R = list(map(int,input().split()))

q = [[] for _ in range(n)]
for _ in range(k):
    x, y = map(int,input().split())
    q[min(x, y)-1].append(max(x, y)-1)
    q[max(x, y)-1].append(min(x, y)-1)

cnt = Counter(R)

keys = list(cnt.keys())
keys.sort(reverse=True)

#print(keys)

d = {}
p = 0
for key in keys:
    d[key] = n-cnt[key]
    d[key] -= p
    p += cnt[key]

ans = [0]*n
for i in range(n):
    ans[i] = d[R[i]]
    for j in q[i]:
        if R[i] > R[j]:
            ans[i] -= 1
            
print(*ans)
```
