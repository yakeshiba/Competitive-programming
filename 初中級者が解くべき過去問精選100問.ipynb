# 全探索：全列挙 [4/4]

```python
#ITP1_7_B - How Many Ways?
ns = []
xs = []
while True:
    n,x = map(int,input().split())
    if n == 0 and x == 0:
        break
    ns.append(n)
    xs.append(x)
ans = []
for i in range(len(ns)):
    tmp = 0
    for a in range(1,ns[i]+1):
        for b in range(a,ns[i]+1):
            for c in range(b,ns[i]+1):
                if a+b+c == xs[i]:
                    if a!=b and b!=c and c!= a:
                        tmp += 1
    ans.append(tmp)
print(*ans,sep="\n")
```

```python
#AtCoder Beginner Contest 106 B - 105
N = int(input())

def make_divisors(n):
    from collections import deque
    divisors = deque([])
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            divisors.append(i)
            if i != n // i:
                divisors.append(n//i)
    lst_divisors = list(divisors)
    lst_divisors.sort()
    return lst_divisors

ans = 0
for i in range(1,-(-N//2)+1):
    now = 2*i-1
    if len(make_divisors(now)) == 8:
        ans += 1

print(ans)
```

```python
#パ研杯2019 C - カラオケ
import itertools

N,M = map(int,input().split())

A = [[0] for _ in range(N)]
for i in range(N):
    A[i] = list(map(int,input().split()))

songs = [i for i in range(M)]

song_choice = itertools.combinations(songs, 2)

ans = 0
for song1, song2 in song_choice:
    group = 0
    for student in range(N):
        group += max(A[student][song1],A[student][song2])
    ans = max(group,ans)

print(ans)
```

# 全探索：ビット全探索 [5/5]

```python
#ALDS_5_A - 総当たり
n = int(input())
A = list(map(int,input().split()))
q = int(input())
M = list(map(int,input().split()))

possible = []

for status in range(1<<n):
    tmp = 0
    for mask in range(n):
        if (1<<mask)&status:
            tmp += A[mask]
    possible.append(tmp)
    
possible = set(possible)    
for m in M:
    if m in possible:
        print('yes')
    else:
        print('no')
```

```python
#AtCoder Beginner Contest 128 C - Switches
N,M = map(int,input().split())

related_switch = [[] for _ in range(M)]
for i in range(M):
    s = list(map(lambda x: int(x)-1,input().split()))
    related_switch[i] = s[1:]

p = list(map(int,input().split()))

ans = 0
for status in range(1<<N):#スイッチのon/offの状態
    on = []#onになってるスイッチの番号
    for mask in range(N):
        if (1<<mask)&status:
            on.append(mask)
    for i,s in enumerate(related_switch):
        if len(set(s)&set(on))%2 != p[i]:#積集合
            break
    else:
        ans += 1

print(ans)
```

```python
#AtCoder Beginner Contest 002 D - 派閥
import itertools
N,M = map(int,input().split())

edge = [[0]*N for _ in range(N)]
for i in range(M):
    x,y = map(int,input().split())
    edge[x-1][y-1] = 1
    edge[y-1][x-1] = 1

ans = 1
for status in range(1<<N):#誰が所属しているかを表す状態
    group = []
    for mask in range(N):
        if (1<<mask)&status:
            group.append(mask)
    for p1, p2 in itertools.combinations(group, 2):
        if edge[p1][p2] == 0:
            break
    else:
        ans = max(ans,len(group))
print(ans)
```

```python
#JOI 2008 予選 5 - おせんべい
#numpy sum max axis
import sys
import numpy as np
import copy

input=sys.stdin.readline

def main(args):
    gyou, retsu = map(int,input().split())
    
    senbei = np.array([list(input().split()) for _ in range(gyou)]) == "1"
    
    ans = 0
    for status in range(1<<gyou):
        senbei_tmp = copy.deepcopy(senbei)
        for mask in range(gyou):
            if (1<<mask)&status:
                senbei_tmp[mask] = np.logical_not(senbei_tmp[mask])
        
        white = np.sum(senbei_tmp, axis=0)
        tmp = np.sum(np.max([white,gyou-white], axis=0))
        
        ans = max(ans,tmp)
    
    print(ans)

if __name__ == '__main__':
    main(sys.argv[1:])
```

```python
#Square869120Contest #4 B - Buildings are Colorful!　
N,K = map(int,input().split())
A = list(map(int,input().split()))

ans = float('inf')
for status in range(1<<N):
    cnt = 0
    tmp = 0
    std = 0
    for mask in range(N):
        if (1<<mask)&status:
            cnt += 1
            if std >= A[mask]:
                tmp += std - A[mask] + 1
                std += 1
        
        std = max(std,A[mask])
    if cnt >= K:
        ans = min(ans,tmp)
    
print(ans)
```

# 二分探索

```python
#ALDS_4_B - 二分探索
def binary_s(lst, value):
    left = 0
    right = len(lst)-1
    while left <= right:
        mid = (left + right)//2
        if lst[mid] == value:
            return mid
        elif lst[mid] < value:
            left = mid + 1
        else:
            right = mid -1
    return -1

n = int(input())
S = list(map(int,input().split()))
q = int(input())
T = list(map(int,input().split()))

ans = 0
for t in T:
    if binary_s(S, t) >= 0:
        ans += 1

print(ans)
```

# ダイクストラ法
### N頂点 M辺のグラフにおける頂点 rから各頂点への最短経路長を、O(MlogN)で計算するアルゴリズム
### いつもの距離求解にheapqを使っているだけ

```python
#GRL_1_A - 単一始点最短経路
from heapq import heappush, heappop

v,e,r = map(int,input().split())

edge = [[] for _ in range(v)]

for _ in range(e):
    s,t,d = map(int,input().split())
    edge[s].append((t, d))

cost = [float('inf')]*v
cost[r] = 0

q = [(r, 0)]
while q:
    now, distance = heappop(q)
    #print(q,now,distance)
    
    for nxt, newcost in edge[now]:
        if distance + newcost < cost[nxt] :
            cost[nxt] = distance + newcost
            heappush(q, (nxt, distance + newcost))

for i in range(v):
    if cost[i] == float('inf'):
        print('INF')
    else:
        print(cost[i])
```

# ワ―シャルフロイド法
### N頂点 M辺のグラフにおける全頂点対間の最短経路長を、O(N^3)で計算するアルゴリズム

```python
#GRL_1_C - 全点対間最短経路
#有向グラフのとき
def warshallFloyd(V, dp):
    for k in range(V):
        for i in range(V):
            for j in range(V):
                dp[i][j] = min(dp[i][j], dp[i][k] + dp[k][j])

V, E = map(int,input().split())
dp = [[float('inf')]*V for _ in range(V)]
for i in range(V):
    dp[i][i] = 0
for _ in range(E):
    f, t, d = map(int,input().split())
    dp[f][t] = d
warshallFloyd(V, dp)

ncircle = False
for i in range(V):
    if dp[i][i] < 0:#負経路＝ある点から出て自分に帰ってくる経路＝点iから点i
        ncircle = True
if ncircle:
    print('NEGATIVE CYCLE')
else:
    for d in dp:
        print(' '.join(map(str, d)).replace('inf', 'INF'))
```

# 高速なべき乗計算

```python
#Square869120Contest #1 E - 散歩
from itertools import accumulate
MOD = 10**9+7
N,Q = map(int,input().split())
A = list(map(int,input().split()))
C = list(map(int,input().split()))
distance = [0]*(N-1)
for i in range(N-1):
    distance[i] = pow(A[i],A[i+1],MOD)

#print(distance)
cumsum = [0]+list(accumulate(distance))
#print(cumsum)
ans = 0
now = 0
for c in C:
    ans += abs(cumsum[c-1] - cumsum[now])%MOD
    now = c-1
    #print(ans)
ans += abs(cumsum[0] - cumsum[now])%MOD
print(ans%MOD)
```

# 逆元を使う問題 [1/4]

```python
#AtCoder Beginner Contest 034 C - 経路
mod = 10**9+7

def modInverse(a,p):
    # Fermat's little theorem, a**(p-1) = 1 mod p
    return pow(a,p-2,p)

def nCr(n,r,p=mod):
    num = 1
    den = 1
    for i in range(r):
        num = (num*(n-i)) % p
        den = (den*(i+1)) % p
    return num*modInverse(den,p)% p
    
W,H = map(int,input().split())
print(nCr(W+H-2,H-1))
```

# 累積和 [2/5]

```python
#全国統一プログラミング王決定戦本戦 A - Abundant Resources
import numpy as np

N = int(input())
A = np.array([0] + list(map(int,input().split())))

cumsum = np.cumsum(A)

for i in range(1,N+1):
    #「範囲iの和」の内の最大値を求める
    V = cumsum[i:] - cumsum[:-i]#[:-i]とすると，「最後尾からi個を除いた部分」を切り取れる
    print(np.max(V))
```
