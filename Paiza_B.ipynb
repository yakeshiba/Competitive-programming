```python
#B051
N = int(input())
tmp = []
ans = []

appeared = []
for j in range(N**2):
    appeared.append(j+1)

sum_flag = False#何度も和の計算をしなくていいように
row_sum = None

for i in range(N):
    row = list(map(int,input().split()))
    for num in row:#一度現れた数字を削除
        if num != 0:
            appeared.remove(num)
    if 0  not in row:#0が含まれていないとき
        ans.append(row)
        if not sum_flag:
            row_sum = sum(row)#魔法陣に一意の和を設定
            sum_flag = True#以後用無し
    else:#0が含まれるとき
        if not sum_flag:#１行目に0が含まれるせいで，まだ和が分かってないとき
            tmp.append(i)#後で処理できるよう，行番号を記録
            ans.append(row)
        else:
            if row.count(0) == 1:#0が１個の時
                replace_num = row_sum - sum(row)
                row[row.index(0)] = replace_num
                ans.append(row)
            else:#2個の時
                tmp.append(i)#後で処理
                ans.append(row)

if len(tmp) == 2:#後回しにした分,1行に０が１個ずつのとき
    for tmp_num in tmp:
        ans[tmp_num][ans[tmp_num].index(0)] = row_sum - sum(ans[tmp_num])
    for l in range(N):
        print(*ans[l])
elif len(tmp) == 1:
    if ans[tmp[0]].count(0) == 1:#最初の１行に０が１個あっただけのとき
        ans[tmp[0]][ans[tmp[0]].index(0)] = row_sum - sum(ans[tmp[0]])
        for l in range(N):
            print(*ans[l])
    else:#１行に２個０が含まれるとき
        zeros = [i for i, x in enumerate(ans[tmp[0]]) if x == 0]#0の位置を取得
        #ケース１
        ans[tmp[0]][zeros[0]] = appeared[0]
        ans[tmp[0]][zeros[1]] = appeared[1]
        conf = 0
        for k in range(N):
            conf += ans[k][zeros[0]]
        if conf == row_sum:
            for l in range(N):
                print(*ans[l])
        else:
            ans[tmp[0]][zeros[0]] = appeared[1]
            ans[tmp[0]][zeros[1]] = appeared[0]
            for l in range(N):
                print(*ans[l])
else:
    for l in range(N):
        print(*ans[l])
```

```python
#B031
import numpy as np
N = int(input())
S = np.array(list(input())) == 'b'
#print(S)

if S[0] == 0 and S[-1] == 0:
    print(0)
elif S[0] == 1 and S[-1] == 1:
    print(N)
else:
    groups = []
    tmp = 0
    zeros = 0
    for n in S:
        if n == 1:
            if zeros != 0:#直前まで０だったとき
                groups.append(zeros)
                zeros = 0#リセット
            tmp += 1
        elif n == 0:
            if tmp != 0:#直前まで１だったとき
                groups.append(tmp)
                tmp = 0#リセット
            zeros -= 1

    #終端の処理
    if tmp != 0:
        groups.append(tmp)
    if zeros != 0:
        groups.append(zeros)
    #print(groups)
    
    if len(groups) == 2:
        print(max(groups))
    else:
        while True:
            for i in range(1,len(groups)-1):#両端以外の符号の反転
                groups[i] = -groups[i]
    
            #隣り合ってて符号が一緒のところを結合(該当するのは反転部分の両端のみ)
            tmp_list = []
            for i in range(1,len(groups)-1):
                if i == 1:
                    tmp_list.append(groups[0]+groups[1])
                elif i == len(groups)-2:
                    tmp_list.append(groups[len(groups)-2]+groups[len(groups)-1])
                else:
                    tmp_list.append(groups[i])
            groups = tmp_list
            #print(groups)
            if len(groups) == 2:
                break
        print(max(groups))
```
