```python
#B
n = int(input())
for _ in range(n):
    s = input()
    okyo = s.find("okyo")
    ech = s.rfind("ech")
    if okyo >= 0 and ech >= 0 and okyo < ech:
        print('Yes')
    else:
        print('No')
```

```python
#C
n, x = map(int,input().split())
A = list(map(int,input().split()))

now = 0
ind = []
for i,a in enumerate(A):
    if a == -1:
        ind.append(i)
        continue
    now ^= a

if len(ind) == 0:
    if now != x:
        print(-1)
    else:
        print(*A)
    exit()

ans = x^now

print(ans)

B = [0]*n
for i in range(n):
    if A[i] == -1:
        continue
    B[i] = A[i]

bit = bin(ans)[2:]
print("0b "+bit)

digit = 0
for i in ind:
    tmp = 0
    for d in range(digit,len(bit)):
        if bit[d] == "0":
            digit += 1
            continue
        p = int(bit[d])*2**(len(bit)-1-d)
        if tmp + p <= x:
            tmp += p
            digit += 1
        else:
            break
    B[i] = tmp

if digit < len(bit):
    print(-1)
    exit()
else:
    print(*B)

"""
tmp = 0
for b in B:
    tmp ^= b
print(tmp)
print(x)
"""
```

```python
#D
def create_prime_list(limit):
    x = limit**0.5
    primes = []
    #print('x={0}'.format(x))
    nums = [x for x in range(2, limit+1)]
    while nums[0]<=x:
        primes.append(nums[0])
        current_prime = nums[0]
        nums = [x for x in nums if x%current_prime != 0]
    primes.extend(nums)
    return primes

n = int(input())
A = list(map(int,input().split()))

pl = set(create_prime_list(max(A)))

grundy = []

cnt = 0
for a in A:
    if a == 7:
        grundy.append(3)
    elif a-2 in pl:
        grundy.append(2)
    else:
        grundy.append(1)

ans = 0
for g in grundy:
    ans ^= g
    
if not ans:
    print("Ai")
else:
    print("An")
```
