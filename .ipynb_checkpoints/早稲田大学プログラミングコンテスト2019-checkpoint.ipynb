```python
#A
s = list(input())[::-1]
n = len(s)

a_flag = False
for i in range(n):
    if s[i] == "A":
        a_flag = True
        if i+1 <= n-1 and s[i+1] == "W":
            s[i] = "C"
    elif s[i] == "W":
        if a_flag:
            if i == n-1:
                s[i] = "A"
                continue
            if s[i+1] == "W":
                s[i] = "C"
            else:
                s[i] = "A"
        else:
            continue
    else:
        a_flag = False
        
print("".join(s[::-1]))
```

```python
#B
def ans(m):
    if m==6 or m==7:
        return 1
    elif m==8:
        return 2
    elif m == 9:
        return 3
    else:
        return 0

h, w = map(int,input().split())

ok = False
maze = []
for i in range(h):
    t = list(map(int,input().split()))
    if 5 in t:
        ok = True
    maze.append(t)

maxim = 0
for r in maze:
    maxim = max(max(r), maxim)

if not ok:
    print('Yes 0' if maxim == 0 else 'No')
else:
    if h == 1 or w == 1:
        if w == 1:
            t = []
            for r in maze:
                t.append(r[0])
        if t[0] == 5 or t[-1] == 5:
            print('Yes', ans(maxim)+1)
        else:
            inds = [i for i,v in enumerate(t) if v==5]
            a = 100
            for ind in inds:
                left = 0
                right = 0
                for i in range(0,ind):
                    left = max(left, t[i])
                for j in range(ind+1,max(w,h)):
                    right = max(right, t[j])
                a = min(a, ans(left)+ans(right)+1)
            print('Yes', a)

    else:
        print('Yes', ans(maxim)+1)
```
