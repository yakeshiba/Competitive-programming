```python
#ランクB
N, std = map(int,input().split())
As = list(map(int,input().split()))
Q = int(input())
for i in range(Q):
    S, E = map(int,input().split())
    area_average = sum(As[S-1:E])/(E-S+1)
    if area_average < std:
        if int(std - area_average) - (std - area_average) == 0:
            add_num = int(std - area_average)
        else:
            add_num = int(std - area_average) + 1
        for j in range(E-S+1):
            As[S-1 + j] += add_num
print(*As)
```

```python
#B021
N = int(input())
ans = []
es = ["s", "o", "x"]
vowel = ["a", "i", "u", "e", "o"]
for i in range(N):
    w = input()
    if w[-1] in es:#ses,oes,xes
        ans.append(w+"es")
    elif w[-1] == "h":
        if w[-2] == "s" or w[-2] == "c":#ches,shes
            ans.append(w+"es")
        else:#~hs
            ans.append(w+"s")
    elif w[-1] == "f":
        ans.append(w[:-1]+"ves")
    elif w[-1] == "e":
        if w[-2] == "f":
            ans.append(w[:-2]+"ves")
        else:
            ans.append(w+"s")
    elif w[-1] == "y":
        if w[-2] in vowel:
            ans.append(w+"s")
        else:
            ans.append(w[:-1]+"ies")
    else:
        ans.append(w+"s")
for j in range(N):
    print(ans[j])
```

```python
#B055
N, distance = map(int,input().split())
howmuch = []
ans = [0, 0]
for _ in range(N):
    first_d, first_fee, additional_d, additional_fee = map(int,input().split())
    res_d = distance - first_d
    if res_d < 0:
        howmuch.append(first_fee)
    elif res_d == 0:
        howmuch.append(first_fee+additional_fee)
    else:
        have_to_pay = first_fee
        if res_d%additional_d == 0:
            have_to_pay += additional_fee * (res_d//additional_d)
        else:
            have_to_pay += additional_fee * (int(res_d/additional_d)+1)
        howmuch.append(have_to_pay)
ans[0] = min(howmuch)
ans[1] = max(howmuch)
print(*ans)
```

```python
#B051
N = int(input())
tmp = []
ans = []

appeared = []
for j in range(N**2):
    appeared.append(j+1)

sum_flag = False#何度も和の計算をしなくていいように
row_sum = None

for i in range(N):
    row = list(map(int,input().split()))
    for num in row:#一度現れた数字を削除
        if num != 0:
            appeared.remove(num)
    if 0  not in row:#0が含まれていないとき
        ans.append(row)
        if not sum_flag:
            row_sum = sum(row)#魔法陣に一意の和を設定
            sum_flag = True#以後用無し
    else:#0が含まれるとき
        if not sum_flag:#１行目に0が含まれるせいで，まだ和が分かってないとき
            tmp.append(i)#後で処理できるよう，行番号を記録
            ans.append(row)
        else:
            if row.count(0) == 1:#0が１個の時
                replace_num = row_sum - sum(row)
                row[row.index(0)] = replace_num
                ans.append(row)
            else:#2個の時
                tmp.append(i)#後で処理
                ans.append(row)

if len(tmp) == 2:#後回しにした分,1行に０が１個ずつのとき
    for tmp_num in tmp:
        ans[tmp_num][ans[tmp_num].index(0)] = row_sum - sum(ans[tmp_num])
    for l in range(N):
        print(*ans[l])
elif len(tmp) == 1:
    if ans[tmp[0]].count(0) == 1:#最初の１行に０が１個あっただけのとき
        ans[tmp[0]][ans[tmp[0]].index(0)] = row_sum - sum(ans[tmp[0]])
        for l in range(N):
            print(*ans[l])
    else:#１行に２個０が含まれるとき
        zeros = [i for i, x in enumerate(ans[tmp[0]]) if x == 0]#0の位置を取得
        #ケース１
        ans[tmp[0]][zeros[0]] = appeared[0]
        ans[tmp[0]][zeros[1]] = appeared[1]
        conf = 0
        for k in range(N):
            conf += ans[k][zeros[0]]
        if conf == row_sum:
            for l in range(N):
                print(*ans[l])
        else:
            ans[tmp[0]][zeros[0]] = appeared[1]
            ans[tmp[0]][zeros[1]] = appeared[0]
            for l in range(N):
                print(*ans[l])
else:
    for l in range(N):
        print(*ans[l])
```

```python
#B016
w, h, n = map(int,input().split())
x, y = map(int,input().split())#初期値
for i in range(n):
    direction, step = map(str,input().split())
    step = int(step)
    if direction == "U":
        if y+step%h < h:#下に回り込むとき
            y = y + step%h
        else:
            y = y + step%h - h 
    elif direction == "D":
        if y-step%h >= 0:
            y = y-step%h
        else:
            y = h - abs(y-step%h)
    elif direction == "R":
        if x+step%w < w:
            x = x + step%w
        else:
            x = x + step%w - w 
    else:
        if x-step%w >= 0:
            x = x-step%w
        else:
            x = w - abs(x-step%w)
print(x,y)
```

```python
#B043 ねずみ小僧
H,W = map(int,input().split())
h0, w0 = map(int,input().split())
maze = []
#進行方向
direction = "N"
#d = ["N", "E", "W", "S"]

for i in range(H):
    row = list(input())
    maze.append(row)

#初期値
if maze[h0-1][w0-1] == ".":
    maze[h0-1][w0-1] = "*"
    direction = "E"
    goto = [h0-1,w0]
else:
    maze[h0-1][w0-1] = "."
    direction = "W"
    goto = [h0-1,w0-2]

flag = True#町の中にいるか
while flag:
    if goto[0] <= -1 or goto[0] >= H or goto[1] <= -1 or goto[1] >= W:
        flag = False
        break
    if direction == "N":
        if maze[goto[0]][goto[1]] == ".":
            maze[goto[0]][goto[1]] = "*"
            direction = "E"
            goto[1] += 1
        else:
            maze[goto[0]][goto[1]] = "."
            direction = "W"
            goto[1] -= 1
    elif direction == "E":
        if maze[goto[0]][goto[1]] == ".":
            maze[goto[0]][goto[1]] = "*"
            direction = "S"
            goto[0] += 1
        else:
            maze[goto[0]][goto[1]] = "."
            direction = "N"
            goto[0] -= 1
    elif direction == "W":
        if maze[goto[0]][goto[1]] == ".":
            maze[goto[0]][goto[1]] = "*"
            direction = "N"
            goto[0] -= 1
        else:
            maze[goto[0]][goto[1]] = "."
            direction = "S"
            goto[0] += 1
    else:
        if maze[goto[0]][goto[1]] == ".":
            maze[goto[0]][goto[1]] = "*"
            direction = "W"
            goto[1] -= 1
        else:
            maze[goto[0]][goto[1]] = "."
            direction = "E"
            goto[1] += 1
for s in maze:
    ans_row = "".join(s)
    print(ans_row)
```

```python
#B033 テーブルジェネレーター
W = int(input())
header = list(map(str,input().split()))
header_len = []
for s in header:
    header_len.append(len(s))
H = int(input())
contents = []
contents_is_longer = False
for i in range(H):
    tmp = list(map(str,input().split()))
    contents.append(tmp)
    for t in range(W):
        if len(tmp[t]) > header_len[t]:
            header_len[t] = len(tmp[t])
            if t == 0:
                contents_is_longer = True
#print(contents)
#print(["| "+ contents[0][0]+" "*(header_len[0]-len(contents[0][0]))+" |"])
#１行目
if contents_is_longer:
    ans = [["| "+header[0]+" "*(header_len[0]-len(header[0]))+" |"]]
else:
    ans = [["| "+header[0]+" |"]]
for j in range(W-1):
    row1 = " " + header[j+1]+" "*(header_len[j+1]-len(header[j+1]))+" |"
    ans[0].append(row1)
#２行目
ans.append(["|"+"-"*(header_len[0]+2)+"|"])
for k in range(W-1):
    row2 = "-"*(header_len[k+1]+2)+"|"
    ans[1].append(row2)
#以降
for l in range(H):
    ans.append(["| "+ contents[l][0]+" "*(header_len[0]-len(contents[l][0]))+" |"])
    for m in range(W-1):
        rowm = " "+contents[l][m+1]+" "*(header_len[m+1]-len(contents[l][m+1]))+" |"
        ans[l+2].append(rowm)
for n in ans:
    ans_row = "".join(n)
    print(ans_row)
```

```python
#B034
sx, sy = map(int,input().split())
df, dr, db, dl = map(int,input().split())
N = int(input())
direction = ["F","R","B","L"]
state = "F"
for i in range(N):
    e, c = map(str,input().split())
    if e == "m":#移動
        if c == "F":#正面へ
            if state == "F":#上向き
                sy += df
            elif state == "R":
                sx += df
            elif state == "B":
                sy -= df
            else:
                sx -= df
        elif c == "R":
            if state == "F":
                sx += dr
            elif state == "R":
                sy -= dr
            elif state == "B":
                sx -= dr
            else:
                sy += dr
        elif c == "B":
            if state == "F":
                sy -= db
            elif state == "R":
                sx -= db
            elif state == "B":
                sy += db
            else:
                sx += db
        else:
            if state == "F":
                sx -= dl
            elif state == "R":
                sy += dl
            elif state == "B":
                sx += dl
            else:
                sy -= dl
    else:#ターン
        if c == "L":
            index = direction.index(state)
            if index == 0:
                index = 4
            state = direction[index-1]
        elif c == "B":
            index = direction.index(state)
            if index <= 1:
                index = 4 + index
            state = direction[index-2]
        else:
            index = direction.index(state)
            if index == 3:
                index = -1
            state = direction[index+1]
print(sx,sy)
```
