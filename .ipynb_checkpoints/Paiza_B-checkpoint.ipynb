```python
#ランクB
N, std = map(int,input().split())
As = list(map(int,input().split()))
Q = int(input())
for i in range(Q):
    S, E = map(int,input().split())
    area_average = sum(As[S-1:E])/(E-S+1)
    if area_average < std:
        if int(std - area_average) - (std - area_average) == 0:
            add_num = int(std - area_average)
        else:
            add_num = int(std - area_average) + 1
        for j in range(E-S+1):
            As[S-1 + j] += add_num
print(*As)
```

```python
#B055
N, distance = map(int,input().split())
howmuch = []
ans = [0, 0]
for _ in range(N):
    first_d, first_fee, additional_d, additional_fee = map(int,input().split())
    res_d = distance - first_d
    if res_d < 0:
        howmuch.append(first_fee)
    elif res_d == 0:
        howmuch.append(first_fee+additional_fee)
    else:
        have_to_pay = first_fee
        if res_d%additional_d == 0:
            have_to_pay += additional_fee * (res_d//additional_d)
        else:
            have_to_pay += additional_fee * (int(res_d/additional_d)+1)
        howmuch.append(have_to_pay)
ans[0] = min(howmuch)
ans[1] = max(howmuch)
print(*ans)
```

```python
#B051
N = int(input())
tmp = []
ans = []

appeared = []
for j in range(N**2):
    appeared.append(j+1)

sum_flag = False#何度も和の計算をしなくていいように
row_sum = None

for i in range(N):
    row = list(map(int,input().split()))
    for num in row:#一度現れた数字を削除
        if num != 0:
            appeared.remove(num)
    if 0  not in row:#0が含まれていないとき
        ans.append(row)
        if not sum_flag:
            row_sum = sum(row)#魔法陣に一意の和を設定
            sum_flag = True#以後用無し
    else:#0が含まれるとき
        if not sum_flag:#１行目に0が含まれるせいで，まだ和が分かってないとき
            tmp.append(i)#後で処理できるよう，行番号を記録
            ans.append(row)
        else:
            if row.count(0) == 1:#0が１個の時
                replace_num = row_sum - sum(row)
                row[row.index(0)] = replace_num
                ans.append(row)
            else:#2個の時
                tmp.append(i)#後で処理
                ans.append(row)

if len(tmp) == 2:#後回しにした分,1行に０が１個ずつのとき
    for tmp_num in tmp:
        ans[tmp_num][ans[tmp_num].index(0)] = row_sum - sum(ans[tmp_num])
    for l in range(N):
        print(*ans[l])
elif len(tmp) == 1:
    if ans[tmp[0]].count(0) == 1:#最初の１行に０が１個あっただけのとき
        ans[tmp[0]][ans[tmp[0]].index(0)] = row_sum - sum(ans[tmp[0]])
        for l in range(N):
            print(*ans[l])
    else:#１行に２個０が含まれるとき
        zeros = [i for i, x in enumerate(ans[tmp[0]]) if x == 0]#0の位置を取得
        #ケース１
        ans[tmp[0]][zeros[0]] = appeared[0]
        ans[tmp[0]][zeros[1]] = appeared[1]
        conf = 0
        for k in range(N):
            conf += ans[k][zeros[0]]
        if conf == row_sum:
            for l in range(N):
                print(*ans[l])
        else:
            ans[tmp[0]][zeros[0]] = appeared[1]
            ans[tmp[0]][zeros[1]] = appeared[0]
            for l in range(N):
                print(*ans[l])
else:
    for l in range(N):
        print(*ans[l])
```

```python
#B016
w, h, n = map(int,input().split())
x, y = map(int,input().split())#初期値
for i in range(n):
    direction, step = map(str,input().split())
    step = int(step)
    if direction == "U":
        if y+step%h < h:#下に回り込むとき
            y = y + step%h
        else:
            y = y + step%h - h 
    elif direction == "D":
        if y-step%h >= 0:
            y = y-step%h
        else:
            y = h - abs(y-step%h)
    elif direction == "R":
        if x+step%w < w:
            x = x + step%w
        else:
            x = x + step%w - w 
    else:
        if x-step%w >= 0:
            x = x-step%w
        else:
            x = w - abs(x-step%w)
print(x,y)
```

```python
#B033 テーブルジェネレーター
W = int(input())
header = list(map(str,input().split()))
header_len = []
for s in header:
    header_len.append(len(s))
H = int(input())
contents = []
contents_is_longer = False
for i in range(H):
    tmp = list(map(str,input().split()))
    contents.append(tmp)
    for t in range(W):
        if len(tmp[t]) > header_len[t]:
            header_len[t] = len(tmp[t])
            if t == 0:
                contents_is_longer = True
#print(contents)
#print(["| "+ contents[0][0]+" "*(header_len[0]-len(contents[0][0]))+" |"])
#１行目
if contents_is_longer:
    ans = [["| "+header[0]+" "*(header_len[0]-len(header[0]))+" |"]]
else:
    ans = [["| "+header[0]+" |"]]
for j in range(W-1):
    row1 = " " + header[j+1]+" "*(header_len[j+1]-len(header[j+1]))+" |"
    ans[0].append(row1)
#２行目
ans.append(["|"+"-"*(header_len[0]+2)+"|"])
for k in range(W-1):
    row2 = "-"*(header_len[k+1]+2)+"|"
    ans[1].append(row2)
#以降
for l in range(H):
    ans.append(["| "+ contents[l][0]+" "*(header_len[0]-len(contents[l][0]))+" |"])
    for m in range(W-1):
        rowm = " "+contents[l][m+1]+" "*(header_len[m+1]-len(contents[l][m+1]))+" |"
        ans[l+2].append(rowm)
for n in ans:
    ans_row = "".join(n)
    print(ans_row)
```

```python
#B034
sx, sy = map(int,input().split())
df, dr, db, dl = map(int,input().split())
N = int(input())
direction = ["F","R","B","L"]
state = "F"
for i in range(N):
    e, c = map(str,input().split())
    if e == "m":#移動
        if c == "F":#正面へ
            if state == "F":#上向き
                sy += df
            elif state == "R":
                sx += df
            elif state == "B":
                sy -= df
            else:
                sx -= df
        elif c == "R":
            if state == "F":
                sx += dr
            elif state == "R":
                sy -= dr
            elif state == "B":
                sx -= dr
            else:
                sy += dr
        elif c == "B":
            if state == "F":
                sy -= db
            elif state == "R":
                sx -= db
            elif state == "B":
                sy += db
            else:
                sx += db
        else:
            if state == "F":
                sx -= dl
            elif state == "R":
                sy += dl
            elif state == "B":
                sx += dl
            else:
                sy -= dl
    else:#ターン
        if c == "L":
            index = direction.index(state)
            if index == 0:
                index = 4
            state = direction[index-1]
        elif c == "B":
            index = direction.index(state)
            if index <= 1:
                index = 4 + index
            state = direction[index-2]
        else:
            index = direction.index(state)
            if index == 3:
                index = -1
            state = direction[index+1]
print(sx,sy)
```

```python
#B030
H, W = map(int,input().split())
maze = []
for _ in range(H):
    g = input()
    maze.append(g)
sx, sy = map(int,input().split())
N = int(input())
d = []
for _ in range(N):
    d.append(input())
    
now = (sy-1,sx-1)

for j in range(N):
    way = d[j]
    if way == "L":
        #print("L")
        for i in range(1,now[1]+1):
            if now[1] - i == 0 and maze[now[0]][now[1]-i]:
                now = (now[0],0)
                break
            if maze[now[0]][now[1]-i] == "#" and now[1]-i >= 0:
                pass
            else:
                if now[1]-i < 0:
                    now = (now[0],0)
                else:
                    now = (now[0],now[1]-i)
                break
    elif way == "D":
        #print("D")
        for i in range(1,H-now[0]):
            if now[0]+i == H-1 and maze[now[0]+i][now[1]] == "#":
                now = (H-1,now[1])
                break
            if maze[now[0]+i][now[1]] == "#" and now[0]+i <= H-1:
                pass
            else:
                if now[0]+i > H-1:
                    now = (H-1,now[1])
                else:
                    now = (now[0]+i,now[1])
                break
    elif way == "U":
        #print("U")
        for i in range(1,now[0]+1):
            if now[0]-i == 0 and maze[now[0]-i][now[1]] == "#":
                now = (0,now[1])
                break
            if maze[now[0]-i][now[1]] == "#" and now[0]-i >= 0:
                pass
            else:
                if now[0]-i < 0:
                    now = (0,now[1])
                else:
                    now = (now[0]-i,now[1])
                break
    else:
        #print("R")
        for i in range(1,W-now[1]):
            if now[1]+i == W-1 and  maze[now[0]][now[1]+i] == "#":
                now = (now[0],W-1)
                break
            if maze[now[0]][now[1]+i] == "#" and now[1]+i <= W-1:
                pass
            else:
                if now[1]+i > W-1:
                    now = (now[0],W-1)
                else:
                    now = (now[0],now[1]+i)
                break
    #print(now[0]+1,now[1]+1)
print(now[1]+1,now[0]+1)
```

```python
#B054
PAIZA = {"A":0,"B":1,"C":2,"D":3,"E":4}
s1,s2 = map(str,input().split())
ans = []
l1 = len(s1)
l2 = len(s2)

up = 0#繰り上がりの初期化
common_range = min(l1,l2)
for i in range(1,common_range+1):
    s1_last = s1[-i]
    s2_last = s2[-i]

    num = PAIZA[s1_last] + PAIZA[s2_last]+up
    
    mod = num%5
    ans.append(mod)#筆算のその桁
    
    up = num//5#繰り上がりの有無

    
if l1 >= l2:#s1の方が長いとき
    for i in range(l1-common_range):
        now = s1[-(common_range+1+i)]
        num = PAIZA[now] + up
        mod = num%5
        ans.append(mod)
        up = num//5
else:
    for i in range(l2-common_range):
        now = s2[-(common_range+1+i)]
        num = PAIZA[now] + up
        mod = num%5
        ans.append(mod)
        up = num//5
if up:
    ans.append(1)
ans = ans[::-1]
ans_str = []
for s in ans:
    key = [k for k, v in PAIZA.items() if v == s][0]
    ans_str.append(key)
print("".join(ans_str))
```

```python
#B048
import copy
N, M = map(int,input().split())
bingo = []
opened = [[0]*N for _ in range(N)]
for _ in range(N):
    row = list(map(int,input().split()))
    bingo.append(row)
numbers = []
for _ in range(M-1):
    numbers.append(int(input()))
for k in range(M-1):
    flag_find = False#その番号が見つかったか
    number = numbers[k]
    for i in range(N):
        for j in range(N):
            if bingo[i][j] == number:
                opened[i][j] = 1
                flag_find = True
                break
        if flag_find:#番号が見つかってたら次の番号の探索へ
            break
ans_point = []
#開いてないとこを１マスずつ開けて確認
for i in range(N):
    last_test = copy.deepcopy(opened)
    for j in range(N):
        if last_test[i][j] == 1:
            continue
        else:
            #print(bingo[i][j])
            point = 0#その状態でのビンゴ数を格納
            last_test[i][j] = 1#最後の一つを開ける
            for k in range(N):#横のビンゴ
                if sum(last_test[k]) == N:
                    point += 1
            #print(point)
            for k in range(N):#縦のビンゴ kは列番号
                tmp_col = 0
                for l in range(N):#lは行番号
                    tmp_col += last_test[l][k]
                if tmp_col == N:
                    point += 1
            #print(point)
            #斜めのビンゴ
            tmp_rd = 0#right_down
            tmp_ld = 0#left_down
            for k in range(N):
                tmp_rd += last_test[k][k]
                tmp_ld += last_test[(N-1)-k][k]
            if tmp_rd == N:
                point += 1
            if tmp_ld == N:
                point += 1
            #print(point)
            #print("##########")
            ans_point.append(point)
            last_test = copy.deepcopy(opened)
print(max(ans_point))
```

```python
#B075
N, M = map(int,input().split())
score = []
for _ in range(N):
    row = list(map(int,input().split()))
    score.append(row)
T = []
for j in range(N):
    T.append([0,j])#(その商品の最終的なスコア,商品番号)
for ind in range(M):#指標ごとの順位を出す
    points = []
    for item in range(N):#商品ごと
        points.append([score[item][ind],item])#(指標，商品番号)

    points.sort(reverse=True)
    #print("points",points)
    kaburi = {}#同じスコアの商品がいくつあるか（タイの分だけ順位が繰り下がるのでタイの商品数を求めなくてはならない）
    for item in range(N):
        if points[item][0] not in kaburi:
            kaburi[points[item][0]] = 1
        else:
            kaburi[points[item][0]] += 1
    #print("kaburi",kaburi)
    
    prev = points[0][0]
    rank = 1
    for i in range(N):
        if i==0:
            T[points[0][1]][0] += N-1
            #print("shohin_number",points[i][1],"plus_points",N-rank)
            continue
        if prev != points[i][0]:#スコアが前と違うとき
            rank += kaburi[prev]
            T[points[i][1]][0] += N-rank
            prev = points[i][0]
        elif prev == points[i][0]:#さっきとスコアが同じとき
            T[points[i][1]][0] += N-rank
        #print("shohin_number",points[i][1],"plus_points",N-rank)
    #print("######################")
#print(T)
T.sort(reverse=True)

#同じスコアの時に商品番号を昇順にしたい
ans_lst = []
ans = {}
for i in range(N):#まだそのスコアの商品が出てきてないなら
    if T[i][0] not in ans:
        ans[T[i][0]] = [T[i][1]]
    else:
        ans[T[i][0]].append(T[i][1])
#print(ans)
for num_lst in ans.values():
    num_lst.sort()
    ans_lst.append(num_lst)
#print(ans_lst)
for s in ans_lst:
    for num in s:
        print(num+1)
```
