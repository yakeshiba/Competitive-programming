```python
n = int(input())

def make_divisors(n):
    from collections import deque
    divisors = deque([])
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            divisors.append(i)
            if i != n // i:
                divisors.append(n//i)
    lst_divisors = list(divisors)
    lst_divisors.sort()
    return lst_divisors

ans = 0
for c in range(1,n+1):
    mk = make_divisors(n-c)
    if len(mk)%2 == 0:
        ans += 2*(len(mk)//2)
    else:
        ans += 2*(len(mk)//2)+1
    
print(ans)
```

```python
###TLE
import itertools
import bisect

mod = 998244353

n, k = map(int,input().split())

area = [0]*(n+1)
dp = [0]*(n+1)
dp[1] = 1

for i in range(k):
    l, r = map(int,input().split())
    area[l] += 1
    if r+1 <= n:
        area[r+1] -= 1
        

area = list(itertools.accumulate(area))    
able = [i for i,v in enumerate(area) if v==1]

for i in range(1,n+1):
    for d in able:
        if i-d <= 0:
            break
        dp[i] = (dp[i]+dp[i-d])%mod

print(dp[n]%mod)
```

dpは0-indexed  
cumsumは1-indexed(cumsum[0]=0, cumsum[1] = dp[0])  
和を求めたい範囲の左端のインデックスをleft,右端をrightとすると  
和S = cumsum[right+1]-cumsum[left]

```python
mod = 998244353

n, k = map(int,input().split())
area = []
for _ in range(k):
    l, r = map(int,input().split())
    area.append((l, r))
    
dp = [0]*n
dp[0] = 1
cs = [0]*(n+1)
cs[1] = 1
for i in range(1,n):
    for l, r in area:
        left = max(0, i-r)
        right = max(0,i-l+1)
        dp[i] += (cs[right]-cs[left])%mod
    cs[i+1] = (cs[i]+dp[i])%mod

#print(cs)
#print(dp)
print(dp[n-1]%mod)
```
