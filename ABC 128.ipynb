```python
#D
from collections import deque
import copy
import sys

def main():
    N, K = map(int,input().split())
    V_original = deque(list(map(int,input().split())))
    ans = 0
    for right in range(K+1):
        if right > N:
            break
        V = copy.deepcopy(V_original)
        value = 0
        minus = deque([])
        for i in range(min(N,right,len(V))):
            j = V.pop()
            if j < 0:
                minus.append(j)
            value += j
    
        for left in range(K-right+1):
            if left > N:
                break
            V_right_ended = copy.deepcopy(V)
            minus_right_ended = copy.deepcopy(minus)
            value_c = value
            for i in range(min(left,N,len(V_right_ended))):
                j = V_right_ended.popleft()
                if j < 0:
                    minus_right_ended.append(j)
                value_c += j
            
            minus_right_ended = list(minus_right_ended)
            minus_right_ended.sort()
    
            for back in range(K-left-right):
                try:
                    value_c -= minus_right_ended[back]
                except IndexError:
                    break
            
            ans = max(ans,value_c)
    print(ans)
if __name__ == '__main__':
    main()
```
