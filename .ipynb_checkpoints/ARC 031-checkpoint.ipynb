```python
#B
import copy
from collections import deque

def check(chizu):
    for r in range(10):
        for c in range(10):
            if chizu[r][c] == "o":
                return False
    return True
    
island = []
land = 0
for i in range(10):
    tmp = list(input())
    island.append(tmp)
    land += tmp.count('o')

dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]

#元から一つかどうか判定
start = None
for row in range(10):
    for col in range(10):
        if island[row][col] == "o":
            start = (row, col)
            break
    if start:
        break
    
ume = copy.deepcopy(island)
stack = deque([start])
ans = "NO"
while stack:
    y, x = stack.pop()
    for dxi, dyi in zip(dx, dy):
        if not (0<=y+dyi<10 and 0<=x+dxi<10) or ume[y+dyi][x+dxi] == "x":
            pass
        else:
            stack.append((y+dyi, x+dxi))
    ume[y][x] = "x"
if check(ume):
    ans = 'YES'

#元から一つじゃないとき
if ans == "NO":
    for row in range(10):
        if ans == 'YES':
            break
        for col in range(10):
            if island[row][col] == "o":
                continue
            ume = copy.deepcopy(island)
            ume[row][col] = "x"
            #深さ優先探索
            stack = deque([(row, col)])
            while stack:
                y, x = stack.pop()
                for dxi, dyi in zip(dx, dy):
                    if not (0<=y+dyi<10 and 0<=x+dxi<10) or ume[y+dyi][x+dxi] == "x":
                        pass
                    else:
                        stack.append((y+dyi, x+dxi))
                ume[y][x] = "x"
            if check(ume):
                ans = "YES"
                break
print(ans)
```
